```python
import requests
from bs4 import BeautifulSoup
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36'}
response= requests.get("https://www.zomato.com/visakhapatnam",headers=headers)
```

```python
content = response.content
soup = BeautifulSoup(content,"html.parser")
container=soup.find("div",attrs={"class":"ui segment row"})
all_areas=container.find_all('a')
areas=[]
for area in all_areas:
    area=area.get_text().replace('\n',' ').strip().split('(')[0].strip()
    areas.append(area)
for i in range(0,len(areas)):
    areas[i]=areas[i].strip().lower()
for i in range(0,len(areas)):
    if areas[i].find(" ")!=-1:
        c=areas[i].find(" ")
        areas[i]=areas[i].replace(" ","-",c)
```

```python
areas
```

```python
s="https://www.zomato.com/visakhapatnam/dine-out-in-siripuram?ref_page=subzone&page=1&nearby=0"
links=[]
links.append(s)
for i in range(0,len(areas)-1):
    if s.find(areas[i])!=-1:
        c=s.find(areas[i])
        s=s.replace(areas[i],areas[i+1],c)
        links.append(s)
```

```python
links
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36'}
responses=[]
```

```python
for i in range(0,len(links)):
    res= requests.get(links[i],headers=headers)
    responses.append(res)
```

```python
len(responses)
soups=[]
```

```python
for i in range(0,len(responses)):
    content = responses[i].content
    soup = BeautifulSoup(content,"html.parser")
    soups.append(soup)
```

```python
pages=[]
for i in range(0,len(soups)):
    container=soups[i].find("div",attrs={"class":"col-l-4 mtop pagination-number"})
    a=int(container.get_text().split('of')[1].strip())
    pages.append(a)
    
    
```

```python
pages
```

```python
l=[]

for i in range(0,len(links)):
    for j in range(pages[i]):
      s=links[i]
      a=s.find(str(j+1))
      s=s.replace(str(j+1),str(j+2),a)
      l.append(s)
    responses=[]
    for i in range(0,len(l)):
      headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36'}
      res= requests.get(l[i],headers=headers)
      responses.append(res)
    list_rest =[]
    total_list=[]
    import pandas
    for i in range(0,len(l)):
      content = responses[i].content
      soup = BeautifulSoup(content,"html.parser")
      container= soup.find_all("div",attrs={"class":"ui cards"})
      multiple_rest_data=container[0].find_all("div",attrs={"card search-snippet-card search-card "})
      individual_rest=multiple_rest_data[0]
      for individual_rest in multiple_rest_data:
        dataframe ={}
        dataframe["rest_name"] = (individual_rest.find("div",attrs={"class":"col-s-12"}).find("a",attrs={"class":"result-title hover_feedback zred bold ln24 fontsize0 "}).text.replace('\n',' '))
        dataframe["rest_address"] = (individual_rest.find("div",attrs={"class":"col-s-12"}).find("a",attrs={"class":"ln24 search-page-text mr10 zblack search_result_subzone left"}).text.replace('\n',' '))
        dataframe["cuisine_type"] = (individual_rest.find("div",attrs={"class":"search-page-text clearfix row"}).find("div",attrs={"class":"clearfix"}).get_text().split(':')[1].strip().split(','))
        dataframe["ratings"]=(individual_rest.find("div",attrs={"class":"ta-right floating search_result_rating col-s-4 clearfix"}).get_text().replace('\n',' ').strip()[0:3])
        list_rest.append(dataframe)
df = pandas.DataFrame(list_rest)
```

```python
df
```

```python
df1=df.copy()
```

```python
ratings=list(df1['ratings'])
```

```python
import numpy as np
for i in range(0,len(ratings)):
    if ratings[i]=="-":
        ratings[i]=np.nan
    if ratings[i]=="NEW":
        ratings[i]=np.nan
```

```python
for i in range(0,len(ratings)):
    ratings[i]=float(ratings[i])
```

```python
df1["ratings"]=ratings
```

```python
df1.dropna(inplace=True)
```

```python
p=0
final=[]
for i in range(0,len(df1)):
    n=df1['cuisine_type'][p]
    for i in range(0,len(n)):
     k={}
     k['cuisine']=(n[i])
     k['ratings']=(df1['ratings'][p])
     k['area']=(df1['rest_address'][p])
     final.append(k)
    p=p+1
```
